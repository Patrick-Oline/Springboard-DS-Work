```python
df['date'] = pd.to_datetime(df['date'])
df['day'] = df['date'].dt.day
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
```

```python
fig, axs = plt.subplots(3, figsize=(16,10), sharex=True)
fig.tight_layout(pad=5.0)

fig.suptitle('Prices From 1967 to 2021')
axs[0].plot(df['year'], df['cost_of_30_second_ad_usd'])
axs[0].set_title('Cost of 30sec SB Ad')

axs[1].plot(df['year'], df['gas_dol/gal'])
axs[1].set_title('Gas (Dollar/Gallon)')

axs[2].plot(df['year'], df['$100_infl_by_year'])
axs[2].set_title('$100 Inflation by Year')



plt.show()
```

I noticed that while inflation had a steady rise from 1967 to 2021, ad time and gas had most sporatic rises over time, but all three did indeed rise throught the years.

```python
col=['cost_of_30_second_ad_usd', 'gas_dol/gal', '$100_infl_by_year']

scaled_data = df[col]

# Scaling the number to graph them together
from sklearn.preprocessing import MinMaxScaler

# define min max scaler
scaler = MinMaxScaler()

scaled_data = pd.DataFrame(scaler.fit_transform(scaled_data), columns = scaled_data.columns)

scaled_data.head()
```

```python
plt.figure(figsize=(16, 9))
ax = plt.axes()

fig.suptitle('Prices From 1967 to 2021')

ax.plot(df['year'], scaled_data['cost_of_30_second_ad_usd'])
ax.plot(df['year'], scaled_data['gas_dol/gal'])
ax.plot(df['year'], scaled_data['$100_infl_by_year'])
ax.scatter(df['year'], scaled_data['cost_of_30_second_ad_usd'])
ax.scatter(df['year'], scaled_data['gas_dol/gal'])
ax.scatter(df['year'], scaled_data['$100_infl_by_year'])


ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend(['Cost of 30sec SB Ad','Gas (Dollar/Gallon)','$100 Inflation by Year'])

plt.show()
```

The overall result confirmed that they all rose by similar rate over time. While the inflation was fairly linear, it is interesting how sporatic the gas prices were and it almost looks as if the ad time is exponential.
